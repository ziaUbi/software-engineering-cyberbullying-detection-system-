```python
import csv

# ===============================
# 1) Leggi END dal file di testo
# ===============================
dev_end_times = []

with open("timestamp_log_0.txt", "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        timestamp, system, status = line.split(",")

        if system == "Development System" and status == "end":
            dev_end_times.append(float(timestamp))

# ======================================
# 2) Leggi START dal file CSV (con header)
# ======================================
classifier_start_times = []

with open("development_log_0.csv", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row["status"] == "start":
            classifier_start_times.append(float(row["timestamp"]))

# =========================
# 3) Calcolo differenze
# =========================
n = min(len(dev_end_times), len(classifier_start_times))

if n == 0:
    raise ValueError("Nessuna coppia valida trovata")

differences = [
    dev_end_times[i] - classifier_start_times[i]
    for i in range(n)
]

import matplotlib.pyplot as plt
plt.plot(range(1, len(differences) + 1), differences, label="Time Differences")
plt.xlabel("Pair Index")
plt.ylabel("Time Difference (seconds)")
plt.title("Time Differences Between Development System End and Classifier Start")
plt.legend()
plt.show()

average_difference = sum(differences) / n

# =========================
# 4) Output
# =========================
print("Differenze (secondi):")
for i, d in enumerate(differences, start=1):
    print(f"{i}: {d:.6f}")

print(f"\nMedia differenze: {average_difference:.6f} s")
```
